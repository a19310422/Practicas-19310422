```python
  import pylab
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
V={'MEX','GDL','TIJ','QRO','MTY','NLU','GAP','TLC','LAP','MZT'}

a=random.randint(1,10)
b=random.randint(1,10)
c=random.randint(1,10)
d=random.randint(1,10)


E=[('MEX','GDL',a),('MEX','MZT',random.randint(1,20)), ('MZT','GAP',random.randint(1,10)), 
   ('GAP','TIJ',random.randint(1,10)), ('TIJ','LAP',random.randint(1,10)), ('LAP','NLU',random.randint(1,10)), 
   ('NLU','QRO',random.randint(1,10)), ('TLC','GDL',random.randint(1,10)), ('MEX','TIJ',random.randint(1,10)), 
   ('MTY','TLC',random.randint(1,10)), ('GDL','MEX',a), ('QRO','GAP',random.randint(1,10)), ('TIJ','MZT',random.randint(1,10)),
   ('MEX','LAP',random.randint(1,10))]

G=nx.DiGraph()
G.add_nodes_from(V)
G.add_weighted_edges_from(E)

pos={'MEX':[1,2], 'GDL':[2,1], 'TIJ':[3,2], 'QRO':[0.5,3], 'MTY':[3.5,3], 'NLU':[0.5,6], 'GAP':[3.5,6], 'TLC':[1,7], 'LAP':[3,7], 'MZT':[2,8]}

weight=nx.get_edge_attributes(G,'weight')
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=weight)
```

```python

best= nx.dijkstra_path(G, 'MTY', 'MZT')
print(best)
```

```python
best_road = nx.DiGraph()

for i in range(len(best)-1):
    best_road.add_edge(best[i], best[i+1])


nx.draw(best_road, node_size = 1000, width = 1, with_labels = True, edge_color ="tab:red",)
plt.show()
```
